```python
from core.audio_io import load_audio, normalize, play_audio
from core.mixing import mix_with_snr
from evaluation.benchmarking import evaluate_method, print_results
from core.visualization import plot_waveforms, plot_spectrogram
from enhancement.PCA import PCADenoiser
import matplotlib.pyplot as plt
import numpy as np
```

```python
clean, fs = load_audio("data/clean/clean_speech.wav")
noise, _ = load_audio("data/noise/white.wav")
noise = normalize(noise)
noisy_input, _ = mix_with_snr(clean, noise, snr_db=5)
```

```python
pca = PCADenoiser(embedding_dim=250, k_components=25)
```

```python
y_cleaned = pca.fit_transform(noisy_input)
y_cleaned = normalize(y_cleaned)
```

```python
play_audio(noisy_input, fs)
```

```python
play_audio(y_cleaned, fs)
```

```python
res_pca = evaluate_method(clean, noisy_input, y_cleaned)
print("\n--- PCA Results ---")
print_results(res_pca)
```
